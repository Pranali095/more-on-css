```python
# Write code to position 5 equal-sized (50 X 50px) boxes A, B, C, D, and E as follows.

<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0"/>
  <style>
    body {
      margin: 0;
      height: 100vh;
      position: relative;
    }

    .box {
      width: 50px;
      height: 50px;
      position: absolute;
      display: flex;
      align-items: center;
      justify-content: center;
      color: white;
      font-weight: bold;
    }

    .boxA {
      margin-top: 25px;
      margin-left: 25px;
      left: 200px;
      top: 200px;
      background-color: red;
    }

    .boxB {
      left: calc(200px + 25px - 30px); /* center of A - 30px */
      top: calc(200px + 25px - 30px);  /* center of A - 30px */
      background-color: green;
    }

    .boxC {
      left: calc(200px + 25px + 30px); /* center of A + 30px */
      top: calc(200px + 25px - 30px);
      background-color: blue;
    }

    .boxD {
      left: calc(200px + 25px - 30px);
      top: calc(200px + 25px + 30px); /* center of A + 30px */
      background-color: orange;
    }

    .boxE {
      left: calc(200px + 25px + 30px);
      top: calc(200px + 25px + 30px);
      background-color: purple;
    }
  </style>
</head>
<body>
  <div class="box boxA">A</div>
  <div class="box boxB">B</div>
  <div class="box boxC">C</div>
  <div class="box boxD">D</div>
  <div class="box boxE">E</div>
</body>
</html>

```

que. xplain the difference between Absolute and Relative positioning.

1. Absolute positioning :-
- Positioned relative to the closest positioned ancestor or the document body if no ancestor is found.
- Yes, it is removed from the document flow, and other elements may shift.

2. Relative positioning :-  
- Positioned relative to its normal location in the flow.
- No, it doesnâ€™t affect the position of other elements.

```python
# Create a card as shown in the picture below. (You can use CSS float property only for layout).
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Float Card</title>
    <style>
        * {
            margin: 0;
            padding: 0;
            box-sizing: border-box;
        }

        body {
            font-family: Arial, sans-serif;
            background-color: #f4f4f4;
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
        }

        .card {
            width: 300px;
            background: white;
            border-radius: 8px;
            overflow: hidden;
            box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.2);
            padding: 15px;
        }

        .card img {
            width: 100%;
            height: auto;
            display: block;
        }

        .card-content {
            float: left;
            width: 100%;
            padding: 10px 0;
        }

        .card h3 {
            margin-bottom: 10px;
            color: #333;
        }

        .card p {
            color: #666;
            font-size: 14px;
        }

        /* Clearing float */
        .clearfix::after {
            content: "";
            display: table;
            clear: both;
        }
    </style>
</head>
<body>

    <div class="card clearfix">
        <img src="https://via.placeholder.com/300x180" alt="Card Image">
        <div class="card-content">
            <h3>Card Title</h3>
            <p>This is a simple description of the card. It is built using the CSS float property.</p>
        </div>
    </div>

</body>
</html>
```

```python
# Create a simple header that sticks to the top of a webpage upon scrolling.
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>Sticky Header</title>
  <style>
    body {
      margin: 0;
      font-family: Arial, sans-serif;
    }

    .header {
      background-color: #333;
      color: white;
      padding: 15px 20px;
      position: sticky;
      top: 0;
      z-index: 1000;
    }

    .content {
      padding: 20px;
    }

    .content p {
      margin: 20px 0;
    }
  </style>
</head>
<body>

  <div class="header">
    <h2>Sticky Header</h2>
  </div>

  <div class="content">
    <!-- Filler content to enable scrolling -->
    <p>Lorem ipsum dolor sit amet...</p>
    <p>Scroll down to see the header stay fixed!</p>
    <!-- Repeat content for demo -->
    <p>Lorem ipsum dolor sit amet...</p>
    <p>Lorem ipsum dolor sit amet...</p>
    <p>Lorem ipsum dolor sit amet...</p>
    <p>Lorem ipsum dolor sit amet...</p>
    <p>Lorem ipsum dolor sit amet...</p>
    <p>Lorem ipsum dolor sit amet...</p>
  </div>

</body>
</html>
```
